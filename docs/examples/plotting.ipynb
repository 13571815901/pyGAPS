# Graphing examples
## Import isotherms

```python
import os
import matplotlib.pyplot as plt
import pygaps

json_path = os.path.join(os.getcwd(), 'data', 'characterisation')
json_file_paths = pygaps.util_get_file_paths(json_path, '.json')
isotherms = []
for filepath in json_file_paths:
    with open(filepath, 'r') as text_file:
        isotherms.append(pygaps.isotherm_from_json(
            text_file.read()))
```

## Isotherm display

```python
isotherm = isotherms[0]
isotherm.print_info()
```

## Isotherm plotting and comparison

```python
legend_list = ['sample_name', 'adsorbate']
```

```python
fig_title = "Regular isotherms colour"
pygaps.plot_iso(isotherms, plot_type='isotherm', branch=['ads'],
                fig_title=fig_title, legend_list=legend_list)
plt.show()
```

```python
fig_title = "Log isotherms colour, legend at bottom"
pygaps.plot_iso(isotherms, plot_type='isotherm', branch=['ads'],
                  logx=True, fig_title=fig_title, legend_list=legend_list, legend_bottom=True)
plt.show()
```

```python
fig_title = "Black and white"
pygaps.plot_iso(isotherms, plot_type='isotherm', branch=['ads', 'des'],
                  color=False, fig_title=fig_title, legend_list=legend_list)
plt.show()
```

```python
fig_title = "Ranges selected"
x_range = (0.2, 0.6)
y1_range = (3, 10)

pygaps.plot_iso(isotherms, plot_type='isotherm', branch=['ads', 'des'],
                  x_range=x_range, y1_range=y1_range, fig_title=fig_title, legend_list=legend_list)
plt.show()
```

```python
fig_title = "Different pressure mode or units"
mode_pressure = 'relative'
unit_loading = 'cm3 STP'

pygaps.plot_iso(isotherms, plot_type='isotherm', branch=['ads'],
                mode_pressure=mode_pressure, unit_loading=unit_loading, 
                fig_title=fig_title, legend_list=legend_list)
plt.show()
```

```python
fig_title = "Only desorption branch"

pygaps.plot_iso(isotherms, plot_type='isotherm', branch=['des'],
                fig_title=fig_title, legend_list=legend_list)
plt.show()
```
